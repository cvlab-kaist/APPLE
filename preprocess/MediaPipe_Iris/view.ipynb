```python
# /mnt/data2/dataset/VGGface2_None_norm_512_true_bygfpgan/n000003/landmark_proto_visualize/0060_01.jpg
from glob import glob
from natsort import natsorted
proto_paths = natsorted(list(glob('/mnt/data2/dataset/VGGface2_None_norm_512_true_bygfpgan/*/iris_vis/*.png')))
import random

```

```python
import os
# Randomly select one path
random_path = random.choice(proto_paths)
# random_path = proto_paths[0]
random_id = random_path.split('/')[-3]
random_base_name = random_path.split('/')[-1].replace('.png', '')
black_path = os.path.join('/mnt/data2/dataset/VGGface2_None_norm_512_true_bygfpgan', random_id, 'iris', f'{random_base_name}.png')
mm_path = os.path.join('/mnt/data2/dataset/VGGface2_None_norm_512_true_bygfpgan', random_id, '3dmm_vis', f'{random_base_name}.jpg')

from PIL import Image
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
img1 = Image.open(random_path)
img2 = Image.open(mm_path)
black = Image.open(black_path)
axs[0].imshow(img1)
axs[0].set_title(f'MediaPipe : {random_id}, {random_base_name}')
axs[0].axis('off')
axs[1].imshow(black)
axs[1].set_title(f'MediaPipe Black : {random_id}, {random_base_name}')
axs[1].axis('off')
axs[2].imshow(img2)
axs[2].set_title(f'3DMM  : {random_id}, {random_base_name}')
axs[2].axis('off')
plt.show()
# --- IGNORE ---
```
